```python
from transformers import GPT2Tokenizer, GPT2LMHeadModel
```

```python
tokenizer = GPT2Tokenizer.from_pretrained('gpt2-large')
model = GPT2LMHeadModel.from_pretrained('gpt2-large', pad_token_id = tokenizer.eos_token_id)
```

```python
tokenizer.eos_token_id
```

```python
tokenizer.decode(tokenizer.eos_token_id)
```

```python
sentence = input("Enter the text Akaaah...:")
```

```python
numeric_ids = tokenizer.encode(sentence, return_tensors = 'pt')
```

```python
numeric_ids
```

```python
tokenizer.decode(numeric_ids[0][0])
```

```python
result = model.generate( numeric_ids, max_length = 50, num_beams = 5, no_repeat_ngram_size = 2, early_stopping = True)
```

```python
result
```

```python
tokenizer.decode( result[0], skip_special_tokens = True)
```
